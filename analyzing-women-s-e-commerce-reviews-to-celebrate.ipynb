# **Analyzing Women's E-commerce Reviews to Celebrate International Women's Day**

# Introduction

March 8th marks International Women's Day, a day dedicated to celebrating the achievements and contributions of women worldwide. To commemorate this day, we will be analyzing a public dataset about women to gain insights into their lives and experiences. We will be using Python to perform the analysis and visualize the data, making it easy for beginners to follow along.

# Dataset

We will be using the "Women's Clothing E-Commerce Reviews" dataset available on Kaggle. This dataset contains customer reviews and ratings of clothing products from an e-commerce website. It includes information about the clothing item, customer demographics, and their reviews and ratings of the product.

**[READ MORE](https://medium.com/@yennhi95zz/celebrating-international-womens-day-analyzing-a-public-dataset-to-appreciate-the-women-in-our-e8fa10d6aea0)**

```python
# Import the necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Read the dataset
data = pd.read_csv('/kaggle/input/womens-e-commerce-clothing-reviews/Womens Clothing E-Commerce Reviews.csv')
data.head()
```

```python
# Visualizing the distribution of ratings
sns.countplot(x='Rating', data=data)
plt.xlabel('Rating')
plt.ylabel('Count')
plt.title('Distribution of Ratings')
plt.show()
```

```python
# Relationship between age and ratings
sns.boxplot(x='Rating', y='Age', data=data)
plt.xlabel('Rating')
plt.ylabel('Age')
plt.title('Relationship between Age and Ratings')
plt.show()
```

# Recommendations
From the dataset, we can draw several conclusions and make recommendations for the e-commerce industry. For example, since the majority of reviews have a rating of 4 or 5, it is essential to focus on providing high-quality products and excellent customer service. Additionally, since there is a slight relationship between age and ratings, it may be helpful to tailor marketing strategies to specific age groups to increase customer satisfaction.
